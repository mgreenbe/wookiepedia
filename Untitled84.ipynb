```python
from bs4 import BeautifulSoup, Comment
from multiprocessing import Pool
import requests
import re
import json
```

```python
PREFIX = "https://starwars.fandom.com"

def get_next_url(soup):
  link_tags = soup.select(".mw-allpages-nav > a")
  link_tag = link_tags[-1]
  if link_tag.text[:9] == "Next page":
    return PREFIX + link_tag.get("href")
  else:
    return None

url = "https://starwars.fandom.com/wiki/Local_Sitemap"
urls = [url]
i = 0
while True:
  if i > 1000:
    break
  i += 1
  print(f"{i}. {url}")
  html = requests.get(url).text
  soup = BeautifulSoup(html, 'html.parser')
  url = get_next_url(soup)
  urls.append(url)
  if url is None:
    break
print(f"{i}. {url}")
```

```python
PREFIX = "https://starwars.fandom.com"
ref = re.compile(r"\[\d+\]")


def scrape_sitemap_page(url):
  html = requests.get(url).text
  soup = BeautifulSoup(html, 'html.parser')
  links = soup.select("ul.mw-allpages-chunk > li > a")
  hrefs = [PREFIX + link.get("href") for link in links]
  return hrefs


def scrape_old(url):
  html = requests.get(url).text
  soup = BeautifulSoup(html, 'html.parser')
  heading = soup.select_one("#firstHeading").get_text().strip()
  content = [p for p in soup.select(".mw-parser-output > p") if p.find("aside") is None]
  # content = soup.select(".mw-parser-output > p")
  text = "\n\n".join([p.get_text().strip() for p in content])
  text = re.sub(ref, "", text)
  return heading, text

def scrape(url):
  html = requests.get(url).text
  soup = BeautifulSoup(html, 'html.parser')
  heading = soup.select_one("#firstHeading").get_text().strip()
  content = soup.select_one(".mw-parser-output")
  for child in content.children:
    if isinstance(child, Comment):
      child.extract()
  text = str(content).strip()
  return heading, text
```

```python
def get_main(html):
  soup = BeautifulSoup(html, 'html.parser')
  return soup.select_one("main")
```

```python
ref_re = re.compile(r"\[\d+\]")
ws_re = re.compile(r"\n\t+")

def scrape(item):
    i, url = item
    print(i, url)
    if i % 100 == 0: print(i)
    html = requests.get(f"https://starwars.fandom.com/{url}").text
    soup = BeautifulSoup(html, 'html.parser')
    main = soup.select_one("main")
    for tag in main.find_all("script"):
        tag.extract()
    for tag in main.find_all(string=lambda text: isinstance(text, Comment)):
        tag.extract()
    nodes = [{"tag": "h1", "text": main.select_one("#firstHeading").get_text().strip()}]
    categories = [tag.get("title") for tag in main.select("ul.categories > li.category.normal > span > a")]
    content = main.select_one(".mw-parser-output")
    infobox = content.select_one(".portable-infobox")
    if infobox is not None:
        infobox.parent.extract()
    for node in content.children:
        if node.name == "p":
            text = node.get_text()
            text = re.sub(ref_re, "", text)
            text = re.sub(ws_re, " ", text)
            nodes.append({"tag": "p", "text": text.strip()})
        if node.name == "h2":
            text = node.select_one("span.mw-headline").get_text().strip()
            nodes.append({"tag": "h2", "text": text})
        if node.name == "h3":
            text = node.select_one("span.mw-headline").get_text().strip()
            nodes.append({"tag": "h3", "text": text})
        if node.name == "h4":
            text = node.select_one("span.mw-headline").get_text().strip()
            nodes.append({"tag": "h4", "text": text})
        if node.name == "h5":
            text = node.select_one("span.mw-headline").get_text().strip()
            nodes.append({"tag": "h5", "text": text})
        if node.name == "h6":
            text = node.select_one("span.mw-headline").get_text().strip()
            nodes.append({"tag": "h6", "text": text})
        with open(f"articles/{i}.json", "w") as f:
            f.write(json.dumps({"url": url, "categories": categories, "nodes":nodes}))
```

```python
pool = Pool(2)
urls = [line.strip() for line in open("urls.txt").readlines()]


pool.map(scrape, enumerate(urls))
```
